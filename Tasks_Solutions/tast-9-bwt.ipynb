## Import Dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

## Exercise: 
Load a dataset (e.g., the Boston Housing dataset from Scikit-Learn) and prepare the data for linear regression (e.g., split into training and testing sets).

```python
# load dataset
data = pd.read_csv('/kaggle/input/bostonhousing/BostonHousing.csv')
```

```python
# copying data so that original data remain unchanged
df = data.copy()
df.head(20)
```

```python
# size of dataset
df.shape
```

```python
# information about dataset
df.info()
```

```python
X = df.drop(columns='medv')
Y = df['medv']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

## Exercise: 
Implement linear regression using Scikit-Learn. Fit the model to the training data.


```python
model = LinearRegression(fit_intercept=True)
model.fit(X_train,Y_train)
```

## Exercise: 
Predict the target variable for the test set using the fitted linear regression model.


```python
# Predict housing prices on the test set
y_pred = model.predict(X_test)
y_pred
```

## Exercise: 
Calculate the Mean Squared Error (MSE) of the linear regression model on the test set.


```python
# Print the mean squared error
mse = mean_squared_error(Y_test, y_pred)
print("Mean Squared Error:", mse)
```

## Exercise: 
Calculate the R-squared value of the linear regression model on the test set.


```python
# Calculate R-squared
r2 = r2_score(Y_test, y_pred)
print(f'R-squared: {r2:.4f}') #format the R-squared value to four decimal
```

## Exercise: 
Plot the regression line along with the actual data points to visually assess the model's performance.

```python
# Plotting the regression line along with the actual data points
plt.figure(figsize=(10, 6))
plt.scatter(Y_test, y_pred, color='blue')
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], color='red', lw=2)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted values')
plt.show()
```

## Exercise: 
Evaluate the model's performance by comparing the predicted values with the actual values. Create a scatter plot of the predicted vs. actual values.


```python
# Scatter plot of the predicted vs. actual values
plt.figure(figsize=(10, 6))
plt.scatter(Y_test, y_pred, color='blue')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted values')
plt.show()
```

## Exercise: 
Interpret the coefficients of the linear regression model. Explain the impact of each feature on the target variable.

```python
# Coefficients of the linear regression model
coefficients = pd.DataFrame(model.coef_, X.columns, columns=['Coefficient'])
coefficients
```

## Interpretation

Each coefficient in the linear regression model quantifies the expected change in median home value (medv) for a one-unit change in the corresponding feature, holding all other features constant. Positive coefficients indicate features that tend to increase home values, while negative coefficients indicate features that tend to decrease home values. These interpretations provide insight into how various factors influence housing prices in the Boston area based on the dataset used.+
