```python
# import necessary dependencies and loading built-in iris dataset

import seaborn as sns
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

## Exercise: 
Calculate the mean, median, and mode of the sepal lengths in the Iris dataset.



```python
# Calculate mean, median, and mode of sepal lengths
sepal_m = iris['sepal_length'].mean()
sepal_med = iris['sepal_length'].median()
sepal_mod = iris['sepal_length'].mode()[0]

print('Mean of Sepal Length:', sepal_m)
print('Median of Sepal Length:',sepal_med)
print('Mode of Sepal Length:', sepal_mod)
```

## Exercise: 
Calculate the variance and standard deviation of the petal widths in the Iris dataset.


```python
# Calculate variance and standard deviation of petal widths
petal_variance = iris['petal_width'].var()
petal_std_dev = iris['petal_width'].std()

print('Variance of petal length:',petal_variance)
print('Standard Deviation of petal length:', petal_std_dev)
```

## Exercise: 
Create a summary table that includes the mean, median, variance, and standard deviation for all numerical features in the dataset.


```python
# Summary table for all numerical features

summary = iris.describe().T
summary
```

## Exercise: 
Define a random variable for the sepal length and calculate the probability distribution of sepal lengths.

```python
# Probability distribution of sepal lengths

sepal_distribution = iris['sepal_length'].value_counts(normalize=True).sort_index()
sepal_distribution
```

## Exercise: 
Plot the probability distribution of sepal lengths using a histogram.

```python
# Plotting the probability distribution 
plt.figure(figsize=(10, 6))
sns.histplot(iris['sepal_length'], kde=True, stat="density")
plt.xlabel('Sepal Length')
plt.ylabel('Density')
plt.title('Probability Distribution of Sepal Lengths')
plt.show()
```

## Exercise: 
Calculate the cumulative distribution function (CDF) for the petal lengths and plot it.


```python
# CDF for petal lengths
plt.figure(figsize=(10, 6))
sns.ecdfplot(data=iris, x='petal_length')
plt.xlabel('Petal Length')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function (CDF) of Petal Lengths')
plt.show()
```

## Exercise: 
Calculate and plot the probability density function (PDF) for sepal width.


```python
# PDF for sepal width

plt.figure(figsize=(10, 6))
sns.kdeplot(data=iris, x='sepal_width')
plt.xlabel('Sepal Width')
plt.ylabel('Density')
plt.title('Probability Density Function (PDF) of Sepal Width')
plt.show()
```

## Exercise: 
Determine the probability of a randomly selected iris flower having a petal length greater than a given value.

```python
# Probability of petal length greater than a given value

given_value = 5.0
prob_value = (iris['petal_length'] > given_value).mean()
prob_value
```

## Exercise: 
Perform a hypothesis test to determine if there is a significant difference in the mean petal length between two species of iris flowers.


```python
# Hypothesis test for mean petal length between two species
species_1 = 'setosa'
species_2 = 'versicolor'

petal_species_1 = iris[iris['species'] == species_1]['petal_length']
petal_species_2 = iris[iris['species'] == species_2]['petal_length']

t_stat, p_value = stats.ttest_ind(petal_species_1, petal_species_2)
t_stat, p_value
```

## Exercise: 
Calculate and interpret the covariance and correlation between sepal length and sepal width.

```python
# Covariance and correlation between sepal length and sepal width
covariance = iris['sepal_length'].cov(iris['sepal_width'])
correlation = iris['sepal_length'].corr(iris['sepal_width'])

covariance, correlation
```
