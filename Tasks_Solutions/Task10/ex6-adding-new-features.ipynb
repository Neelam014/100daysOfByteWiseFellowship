## Loading Data

## Defining New Groups

```python
# Define age groups
age_bins = [29, 39, 49, 59, 69, 79]
age_labels = ['30-39', '40-49', '50-59', '60-69', '70-79']
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)
```

```python
# Define cholesterol level categories
chol_bins = [0, 200, 239, 500]
chol_labels = ['Normal', 'Borderline High', 'High']
df['chol_level'] = pd.cut(df['chol'], bins=chol_bins, labels=chol_labels)
```

```python
# Define blood pressure categories
bp_bins = [0, 120, 129, 139, 180]
bp_labels = ['Normal', 'Elevated', 'High Stage 1', 'High Stage 2']
df['bp_level'] = pd.cut(df['trestbps'], bins=bp_bins, labels=bp_labels)
```

```python
# Define blood pressure categories
bp_bins = [0, 120, 129, 139, 180]
bp_labels = ['Normal', 'Elevated', 'High Stage 1', 'High Stage 2']
df['bp_level'] = pd.cut(df['trestbps'], bins=bp_bins, labels=bp_labels)
```

```python
# Define heart rate categories
thalach_bins = [0, 120, 160, 200, 220]
thalach_labels = ['Low', 'Moderate', 'High', 'Very High']
df['thalach_level'] = pd.cut(df['thalach'], bins=thalach_bins, labels=thalach_labels)
```

```python
# Example combined risk factor score
df['risk_score'] = df['age'] * 0.3 + df['chol'] * 0.3 + df['trestbps'] * 0.2 + df['thalach'] * 0.2
```
