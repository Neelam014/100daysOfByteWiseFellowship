## 1. Handling Missing Data in Titanic Dataset

 - Task: Identify and handle missing values in the Titanic dataset. Experiment with different strategies such as mean/median imputation, mode imputation, and dropping rows/columns.

```python
# Importing dependencies

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
#loading and copying dataset

data = pd.read_csv('/kaggle/input/titanic/train.csv')
df = data.copy()
df.head()
```

```python
# Checking rows and column
df.shape
```

```python
#Checking null values
df.isnull().sum()
```

## Handling Missing Values through Imputation techniques

```python
# Mean imputation
Mean = df.copy()
Mean['Age'].fillna(Mean['Age'].mean())

# Median imputation
Median = df.copy()
Median['Age'].fillna(Median['Age'].median())
```

```python
# Mode imputation
Mode = df.copy()
Mode['Embarked'].fillna(Mode['Embarked'].mode())
```

## Droping 

```python
# Dropping rows with any missing values
Drop_Row = df.dropna()

# Dropping columns with more than a certain percentage of missing values (e.g., 50%)
threshold = 0.5
Drop_Row = df.copy()
Drop_Row = Drop_Row.loc[:, Drop_Row.isnull().mean() < threshold]
```
