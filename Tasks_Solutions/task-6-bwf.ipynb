## Exercise:

Explore the basic statistics of a dataset, such as mean, median, and standard deviation for each feature.

```python
import numpy as np # import numpy library 

data = iris.data
feature_names = iris.feature_names #load features name from iris data 

# Calculate statistics
mean = np.mean(data, axis=0)
median = np.median(data, axis=0)
std_dev = np.std(data, axis=0)

# Print statistics with feature names
for i, feature in enumerate(feature_names):
    print(f"{feature}:")
    print(f"  Mean: {mean[i]}")
    print(f"  Median: {median[i]}")
    print(f"  Standard Deviation: {std_dev[i]}\n")
```

## Exercise:

Create a function that takes a list of numbers and returns a dictionary with the count, mean, median, and standard deviation.

```python
def calculate_statistics(numbers):
    count = len(numbers) #shows number length 
    mean = sum(numbers) / count #mean calculation
    median = sorted(numbers)[count // 2] #median calculation
    std_dev = (sum((x - mean) ** 2 for x in numbers) / count) ** 0.5 # Standard deviation calculation
    return {
        'count': count,
        'mean': mean,
        'median': median,
        'std_dev': std_dev
    }    # return dictionary

numbers = [1, 2, 3, 4, 5]
statistics = calculate_statistics(numbers)
print(statistics)
```

## Exercise:

Implement a simple linear regression model using Scikit-Learn and print the model coefficients

```python
from sklearn.linear_model import LinearRegression

# Create an instance of the LinearRegression model
model = LinearRegression()

# finding the best-fit line for the training data
print(model.fit(X_train, y_train))

# Print the accuracy score of the model on the training data
print('Here is the training accuracy :', model.score(X_train, y_train))

# Print the accuracy score of the model on the test data
print('Here is the test accuracy :', model.score(X_test, y_test))

# Predict the target values for the test data (X_test)
print('Predict the target values :',model.predict(X_test))

# Print the coefficients of the model
print('Print the coefficients of the model :',model.coef_)
```
